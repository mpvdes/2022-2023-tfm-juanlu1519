# ¿Cuántas viviendas de uso turístico hay en la ciudad de Girona?

En este notebook vamos a ver cuántas viviendas de uso turístico hay en Girona. Para el tratamiento de los datos necesitaremos instalar e importar las librerías Numpy y Pandas de Python.

````python
#!pip install numpy
#!pip install pandas

import numpy as np
import pandas as pd
````

Procedemos a leer los archivos donde se representan las actividades económicas de cada año asignandoles el nombre "actXXXX" seguido del año del archivo. Estos archivos los encontramos en la web del Ayuntamiento de Girona dedicada a los datos abiertos, [Girona Open Data](https://www.girona.cat/opendata/dataset/establiments-dedicats-a-activitats-economiques).

````python
act2013 = pd.read_csv('../data/activitats013.csv', sep = ';', encoding='latin-1')
act2013['Year'] = 2013
act2014 = pd.read_csv('../data/activitats2014.csv', sep = ';')
act2014['Year'] = 2014
act2015 = pd.read_csv('../data/activitats-2015.csv', sep = ';')
act2015['Year'] = 2015
act2016 = pd.read_csv('../data/activitats2016.csv', sep = ';')
act2016['Year'] = 2016
act2017 = pd.read_csv('../data/activitats2017.csv', sep = ';')
act2017['Year'] = 2017
act2018 = pd.read_csv('../data/activitats2018.csv', sep = ';')
act2018['Year'] = 2018
act2019 = pd.read_csv('../data/activitats2019.csv', sep = ';')
act2019['Year'] = 2019
act2020 = pd.read_csv('../data/activitats2020.csv', sep = ';')
act2020['Year'] = 2020
act2021 = pd.read_csv('../data/activitats2021.csv', sep = ';')
act2021['Year'] = 2021
act2022 = pd.read_csv('../data/activitats2022.csv', sep = ';')
act2022['Year'] = 2022
````

Hemos añadido la columna 'Year' para, a la hora de concatenar los documentos, saber a qué año pertenece cada actividad económica. Con ```pd.concat``` vamos a unir todas las tablas en una sola.



````python
actividades = pd.concat([act2013, act2014, act2015, act2016, act2017, act2018, act2019, act2020, act2021, act2022])
````

````python
actividades
````

Guardamos la nueva tabla *actividades* en un archivo de formato csv para limpiarlo con OpenRefine.


````python
actividades.to_csv('../data/actividades1322.csv')
````

Al usar la función *Cluster and edit* de OpenRefine podemos ver como las viviendas de uso turístico aparecen en la columna *Descripcio* de diferentes maneras: 

- HABITATGE D'ÚS TURÍSTIC (5100 filas)
- HABITATGE D'US TURÍSTIC (16 filas)
- HABITATGE D'ÙS TURÍSTIC (6 filas)
- HABITATGE D'ÚS TURISTIC (4 filas)
- HABITATGE D'ÚS TURÌSTIC (4 filas)
- HABITATGE DÚS TURÍSTIC (3 filas)
- HABITATGE D'ÚS TÚRISTIC (1 fila)

Unificamos criterios para que todas sean "HABITATGE D'ÚS TURÍSTIC" (vivienda de uso turístico, en castellano) y así nos sea más fácil trabajar con el archivo. Guardamos la tabla modificada y volvemos a abrirla con la librería Pandas de Python.

````python
actividades = pd.read_csv('actividades-economicas/actividades1322refine.csv', low_memory=False)

# Añadimos "low_memory=False" al importar el archivo ya que hay tres columnas que tienen mezcla de strings, enteros, etc. y nos da error.
````

````python
actividades
````

En *vut* guardaremos todas las viviendas de uso turístico que aparecen en nuestro listado. Estas se encuentran en la columna *Descripcio* con el nombre *HABITATGE D'ÚS TURÍSTIC*.

````python
vut = actividades[actividades['Descripcio'] == 'HABITATGE D\'ÚS TURÍSTIC']
vut
````

En *vut_year*, aplicando las funciones de Pandas ```groupby``` y ```count```, guardaremos la lista de viviendas de uso turístico por año para poder hacer un gráfico en el siguiente paso.

````python
vut_year = vut.groupby('Year').Descripcio.count()
vut_year
````

Para hacer un gráfico de barras que represente la evolución del número de viviendas de uso turístico usaremos la librería de Python *matplotlib*. La instalaremos, importaremos y añadiremos los parámetros necesarios para poder dibujar el gráfico. 

````python
#!pip install matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
vut_year.plot(kind = 'bar', color = "#F4A460")
plt.xlabel('Años muestra')
plt.ylabel('Número de viviendas de uso turístico')
plt.title('Comparación del número de viviendas de uso turístico en el periodo 2013-2022')
plt.xticks(rotation=45)
plt.show() #Para mostrar el gráfico
plt.tight_layout()  # Para evitar recorte de etiquetas
#plt.savefig('../images/viviendasturisticas1322.svg', format='svg') #para guardar el gráfico en formato svg
````

Si queremos hacer el gráfico interactivo necesitaremos usar la librería Plotly.
Primero guardaremos en la variable *vut_year_plotly* una tabla con el año (Year) y el número de viviendas (Descripcio) para poder trabajar con ella más adelante.

````python
vut_year_plotly = vut_year.reset_index()
vut_year_plotly
````

Tenemos que instalar e importar la librería Plotly para poder hacer el gráfico interactivo. Seguidamente, añadimos los parámetros necesarios para visualizar nuestros datos.

````python
!pip install plotly
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

trace1 = go.Bar(
    x = vut_year_plotly.Year,
    y = vut_year_plotly.Descripcio,
    name = 'Numero'
)
datas = [trace1]


# Creamos el layout
layout = dict(
            # Título
            title = dict( text = "Evolución del número de viviendas turísticas en Girona en el período 2013-2022"),
      tickmode = 'linear',
      tick0 = 2021,
      dtick = 1
         
   )
            
         
fig = go.Figure(data=datas, layout=layout)
iplot(fig) #Para mostrar el gráfico
fig.write_html('../images/viviendasturisticas1322.html') # Lo guardamos en formato HTML para incrustarlo en la web
````
